## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

*Hint*:
- *Column names should be in lower case*
- *White spaces in column names should be replaced by `_`*
- *`st` could be replaced for `state`*

```python
# Your code here
import pandas as pd

df_insurance = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
# df_insurance

# df_insurance.columns = map(str.lower, df_insurance.columns)
df_insurance.columns = df_insurance.columns.str.lower().str.replace(' ', '_')
df_insurance.rename(columns={'st': 'state'}, inplace=True)
df_insurance.head(50)

```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here
df_insurance["gender"] = df_insurance["gender"].replace(["Femal", "female", "F"], "F", inplace=True)
df_insurance["gender"] = df_insurance["gender"].replace(["Male", "male", "M"], "M", inplace=True)

state_mapping = {'Arizona': 'Arizona', 'Washington': 'Washington','Nevada': 'Nevada','California':'California','Oregon':'Oregon','Cali':'California','AZ':'Arizona','WA':'Washington'}
df_insurance['state'] = df_insurance['state'].map(state_mapping)

education_mapping = {'Master': 'Master', 'Bachelor': 'Bachelor','Bachelors': 'Bachelor','Doctor':'Doctor','High School or Below':'High School or Below','College':'College'}
df_insurance['education'] = df_insurance['education'].map(education_mapping)

df_insurance['customer_lifetime_value'] = df_insurance['customer_lifetime_value'].str.replace('%', '')

vehicle_class_mapping = {"Sports Car": "Luxury", "Luxury SUV": "Luxury","Luxury Car": "Luxury",'Four-Door Car':'Four-Door Car','Two-Door Car':'Two-Door Car','SUV':'SUV'}
df_insurance['vehicle_class'] = df_insurance['vehicle_class'].map(vehicle_class_mapping)

df_insurance.head(50)
```
